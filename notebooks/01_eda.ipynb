# Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", palette="muted", font_scale=1.1)
```

# Load Dataset

```python
df = pd.read_csv("../data/processed/municipalities_connectivity_processed.csv")
df.head()
```

# Basic Overview

```python
df.info()
df.describe(include='all')
```

# Histogram of IBC

```python
plt.figure(figsize=(10,6))
sns.histplot(df["ibc"], bins=40, kde=True)
plt.title("Distribution of IBC Across All Municipalities")
plt.xlabel("IBC")
plt.ylabel("Count")
plt.show()
```

# Boxplot: IBC by Region

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="region", y="ibc", order=["N","NE","CO","SE","S"])
plt.title("IBC by Region")
plt.show()
```

# Distribution of Coverage 4G/5G

```python
plt.figure(figsize=(10,6))
sns.histplot(df["coverage_4g5g"], bins=40, kde=True)
plt.title("Distribution of 4G/5G Coverage")
plt.show()
```

# Boxplot: Coverage by Region

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="region", y="coverage_4g5g", order=["N","NE","CO","SE","S"])
plt.title("4G/5G Coverage by Region")
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(12,8))
corr = df[[
    "ibc", 
    "coverage_4g5g", 
    "density_smp", 
    "density_scm",
    "hhi_smp", 
    "hhi_scm",
    "stations_density"
]].corr()

sns.heatmap(corr, annot=True, cmap="viridis", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

# Pairplot

```python
subset = df[[
    "ibc", 
    "coverage_4g5g", 
    "density_smp", 
    "density_scm", 
    "hhi_smp", 
    "hhi_scm"
]]

sns.pairplot(subset, corner=True)
plt.show()
```

# Summary by Region

```python
df.groupby("region")["ibc"].mean().sort_values()
```
